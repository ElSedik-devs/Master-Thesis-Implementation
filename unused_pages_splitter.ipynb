#Find unused pages in the original train and test dataset

```python
import shutil
from pathlib import Path

ALL_PAGES_DIR  = Path("all_pages")
TRAINVAL_DIR   = Path("sampled_pages")   # output_sampled
TEST_DIR       = Path("sampled_test")    # TEST_IMG_DIR
UNUSED_OUT_DIR = Path("unused_pages")    # new directory

EXTS = {".jpg", ".jpeg", ".png", ".webp", ".tif", ".tiff"}

def list_images(folder: Path):
    return sorted([p for p in folder.iterdir() if p.is_file() and p.suffix.lower() in EXTS])

all_pages  = list_images(ALL_PAGES_DIR)
trainval   = list_images(TRAINVAL_DIR) if TRAINVAL_DIR.exists() else []
test_pages = list_images(TEST_DIR) if TEST_DIR.exists() else []

# Compare by stem: page_0006 (ignores extension differences)
used_stems = {p.stem for p in trainval} | {p.stem for p in test_pages}
unused = [p for p in all_pages if p.stem not in used_stems]

UNUSED_OUT_DIR.mkdir(parents=True, exist_ok=True)

for p in unused:
    shutil.copy2(p, UNUSED_OUT_DIR / p.name)

print(f"Copied {len(unused)} unused images to: {UNUSED_OUT_DIR.resolve()}")
```

```python
print("All pages:", len(all_pages))
print("Train/val pages:", len(trainval))
print("Test pages:", len(test_pages))
print("Used stems:", len(used_stems))
print("Unused pages:", len(unused))
```

```python
import random
import shutil
from pathlib import Path

MOBILE_CANDIDATES_DIR = Path("mobile_candidates")
MOBILE_CANDIDATES_DIR.mkdir(parents=True, exist_ok=True)

random.seed(42)

# pick 40 random unused pages, then sort them by filename/page order
selected = sorted(random.sample(unused, min(40, len(unused))), key=lambda p: p.name)

for p in selected:
    shutil.copy2(p, MOBILE_CANDIDATES_DIR / p.name)

print(f"Copied {len(selected)} random-but-sorted unused pages to: {MOBILE_CANDIDATES_DIR.resolve()}")
print("\nSelected pages:")
for p in selected:
    print(p.name)
```

```python
import random
import shutil
from pathlib import Path

ALL_PAGES_DIR = Path("all_pages")
TRAINVAL_DIR = Path("sampled_pages")
TEST_DIR = Path("sampled_test")
MOBILE_TEST_DIR = Path("mobile_candidates")   
NEW_MOBILE_CANDIDATES_DIR = Path("mobile_train_candidates")

EXTS = {".jpg", ".jpeg", ".png", ".webp", ".tif", ".tiff"}

def list_images(folder: Path):
    if not folder.exists():
        return []
    return sorted([p for p in folder.iterdir() if p.is_file() and p.suffix.lower() in EXTS])

all_pages = list_images(ALL_PAGES_DIR)
trainval = list_images(TRAINVAL_DIR)
test_pages = list_images(TEST_DIR)
mobile_test_pages = list_images(MOBILE_TEST_DIR)

# Exclude everything already used anywhere
used_stems = (
    {p.stem for p in trainval}
    | {p.stem for p in test_pages}
    | {p.stem for p in mobile_test_pages}
)

remaining_unused = [p for p in all_pages if p.stem not in used_stems]

NUM_TO_SELECT = 50
random.seed(42)

selected_new_mobile = sorted(
    random.sample(remaining_unused, min(NUM_TO_SELECT, len(remaining_unused))),
    key=lambda p: p.name
)

NEW_MOBILE_CANDIDATES_DIR.mkdir(parents=True, exist_ok=True)

for p in selected_new_mobile:
    shutil.copy2(p, NEW_MOBILE_CANDIDATES_DIR / p.name)

print(f"All pages: {len(all_pages)}")
print(f"Train/val pages: {len(trainval)}")
print(f"Scanned test pages: {len(test_pages)}")
print(f"Mobile test pages: {len(mobile_test_pages)}")
print(f"Total excluded stems: {len(used_stems)}")
print(f"Remaining unused pages: {len(remaining_unused)}")
print(f"Copied {len(selected_new_mobile)} new mobile-train candidates to: {NEW_MOBILE_CANDIDATES_DIR.resolve()}")

print("\nSelected new mobile-train candidate pages:")
for p in selected_new_mobile:
    print(p.name)
```
